```python
def rotateXY(x, y, angle):
    angle *= np.pi/180
    xp = x * np.cos(angle) - y * np.sin(angle)
    yp = x * np.sin(angle) + y * np.cos(angle)
    return (xp,yp)

def halfEllipse(x,x0,xr,yr,angle):
    xp,yp = rotateXY(x-x0,0,-angle)
    yp = yr * np.real( np.sqrt( 1- (xp/xr)**2 ) )
    [x, y] = rotateXY(xp, yp, angle)
    return (x+x0,y)
```
