```python
import pandas as pd 
import numpy as np
```

```python
df = pd.DataFrame()
L = 10
df['x'] = np.ones(L) * 0.8
df['y'] = np.random.randint(0, 100, L)
df['note'] = ['Hi']*L
df['type'] = ['Kyte']*L
df['tof'] = np.ones(L) * 8.1
df.loc[3, 'note'] = 'Hello'
df
```

```python
def find_changing_column_names(df):
    changing_columns = []
    for c, v in df.iteritems():
        try: uniq = np.unique(v)
        except: uniq = [None]
        if len(uniq) > 1: changing_columns.append(c)
    return changing_columns

def get_changing_columns(df):
    return df[find_changing_column_names(df)]

print(find_changing_column_names(df))
print()
get_changing_columns(df)
```
