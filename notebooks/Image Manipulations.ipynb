```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

# moveImage
a.k.a. image panning 

```python
x = np.arange(100)
y = np.arange(102) 
X, Y = np.meshgrid(x, y) 

image = np.sin(0.1 * X*Y)**2 

plt.imshow(image, origin=0)
plt.colorbar()
```

```python
import scipy.interpolate
def moveImage(image, by=[0,0]): 
    '''
    image : 2D array 
    by = [dX, dY] : how many pixels to move the image by 
    '''
    x, y = np.arange(image.shape[1]), np.arange(image.shape[0])
    f = scipy.interpolate.interp2d(x, y, image, 
                    kind='linear', copy=False, 
                    bounds_error=False, fill_value=None)
    return f(x-by[0], y-by[1])
```

```python
plt.imshow(moveImage(image, [-1.5, 10]), origin=0)
```
